## 1. List all unique cities where customers are located.


```python
query= """ select distinct customer_city from customers """
cur.execute(query)

data= cur.fetchall()
df= pd.DataFrame(data, columns= ['Cities'])
print('Cities where customers are located are:')
df
```

## 2. Count the number of orders placed in 2017.


```python
query= """ select count(order_id) from orders where order_purchase_timestamp = 2017 """
cur.execute(query)

data= cur.fetchall()
print('Total orders plased in 2017 are =', data[0][0])
```

## 3. Find the total sales per category.


```python
query= """ select upper(products.product_category) as category, round(sum(payments.payment_value), 2) as sales
            from products join order_items
            on products.product_id = order_items.product_id
            join payments
            on payments.order_id = order_items.order_id
            group by category """
cur.execute(query)

data= cur.fetchall()
df= pd.DataFrame(data, columns= ['Category', 'Sales'])
print('Total sales per categories are:')
df
```

## 4. Calculate the percentage of orders that were paid in installments.


```python
query= """ select (sum(case when payment_installments >= 1 then 1 else 0 end))/ count(*)*100 as cnt from payments """
cur.execute(query)

data= cur.fetchall()
print('The percentage of orders which are paid via installments is: ', data[0][0], '%')
```

## 5. Count the number of customers from each state. 

```python
query= """ select customer_state, count(customer_id) from customers group by customer_state """
cur.execute(query)

data= cur.fetchall()
df = pd.DataFrame(data, columns= ['State', 'No. of Customers'])
print('The number of customer from each states are: ')
df
```
